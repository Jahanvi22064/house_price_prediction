```python
def mc_remover(data):
    Vif=pd.Series([vif(data.values,i)for i in range(data.shape[1])],index=data.columns)
    if Vif.max()>5:
        print(Vif[Vif==Vif.max()].index[0],'has been removed')
        data=data.drop(columns=[Vif[Vif==Vif.max()].index[0]])
        return data
    else:
        print('No multicollnearity present anymore')
        return data
```

```python
for i in range(7):
    vif_data=mc_remover(vif_data)
vif_data.head()
```

```python
def adj_r2(data_x,data_y,predictions):
    from sklearn.metrics import r2_score
    r=r2_score(data_y,predictions)
    n=len(data_y)
    m=len(data_x.columns)
    adj_r=1-((1-r)*(n-1))/(n-m-1)
    return adj_r
```

```python
adj_r2(x_test,y_test,prdictions)
```

```python
def return_r2(x,y,model):
    model.fit(x,y)
    pred=model.predict(x)
    r2=adj_r2(x,y,pred)
    return r2
```

```python
def forward_selection(model,x,y,max_features):
    f_col=[]
    history=[]
    for i in range(max_features):
        f_col.append('blank')
        tmp=f_col[:]
        rmax=-1
        for var in x.columns:
            tmp[-1]=var
            r2=return_r2(x[tmp],y,model)
            if r2>=rmax:
                rmax=r2
                f_col[-1]=var
                elim=var
        print("selected",f_col[-1],"with new r2 ",rmax)
        history.append(rmax)
    return rmax,history,x[f_col]
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
score,history,sel_data=forward_selection(lr,x_train,y_train,29)
```

```python
def backward_eleimination(model,x,y,min_feature):
    elim_data=x[:]
    r2_history=[]
    base_r2=return_r2(elim_data,y,model)
    print("base_r2 is ",base_r2)
    for i in range(x.shape[1]-min_feature):
        diff=-100
        for var in elim_data.columns:
            tmp =elim_data.drop(columns=[var])
            r2=return_r2(tmp,y,model)
            if(r2-base_r2)>diff:
                diff=r2-base_r2
                eliminate=var
                new_r2=r2
        elim_data.drop(columns=[eliminate],inplace=True)
        r2_history.append(new_r2)
        print(eliminate," eleminated: new adj_r2",new_r2)
        base_r2=new_r2
    return base_r2,r2_history,elim_data
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
score,history,sel_data=backward_eleimination(lr,x_train,y_train,25)
```
